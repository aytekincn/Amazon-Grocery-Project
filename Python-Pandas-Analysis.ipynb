# Data visualisation

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
grocery.plot(x = "Category", y = "Sales")
grocery.plot()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Sales', data=grocery)
plt.xticks(rotation=45)
plt.title('Category Sales')
plt.tight_layout()
plt.show()
```
